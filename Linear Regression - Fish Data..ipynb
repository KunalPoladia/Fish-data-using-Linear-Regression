### Q2: Identifying Outliers & Removing them

```python
# There should be no outliers in the data.
# Check using boxplot.
for a in data.columns:
    data.boxplot(column=a)
    plt.show()
```

```python
# Outliers are detected in Weight,Length1,Length2,Length3.
# Outliers of Weight.
def outlier_det(dataframe):
    Q1 = dataframe.quantile(0.25)
    Q3 = dataframe.quantile(0.75)
    IQR = Q3 - Q1
    upper_end = Q3 + 1.5 * IQR
    lower_end = Q1 - 1.5 * IQR 
    outlier = dataframe[(dataframe > upper_end) | (dataframe < lower_end)]
    return outlier
```

```python
outlier_det(data['Weight'])
```

### Q3:- Building a Regression Model & Regression Equation

```python
from sklearn.model_selection import train_test_split  

#Split the data into test and train
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
print('X_train: ', np.shape(X_train))
print('Y_train: ', np.shape(Y_train))
print('X_test: ', np.shape(X_test))
print('Y_test: ', np.shape(Y_test))
```

```python
from sklearn.linear_model import LinearRegression
#create a model object
lm = LinearRegression()

#train the model object
lm.fit(X_train,Y_train)

#print intercept and coefficients
print(lm.intercept_)
print(lm.coef_)
```

```python
#pair the feature names with the coefficients
print(list(zip(X.columns,lm.coef_)))
```

```python
#predict using the model
Y_pred=lm.predict(X_test)
print(Y_pred)
```

```python
lm.score(X_train,Y_train)
```

```python
new_df=pd.DataFrame()
new_df=X_test

new_df["Actual weights"]=Y_test
new_df["Predicted weights"]=Y_pred
new_df
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np

r2=r2_score(Y_test,Y_pred)
print("R-squared:",r2)

rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print("RMSE:",rmse)

adjusted_r_squared = 1 - (1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print("Adj R-square:",adjusted_r_squared)
```

```python
print(Y_test.min())
print(Y_test.max())
```

```python
new_df["Deviation"]=new_df["Actual weights"]-new_df["Predicted weights"]
new_df.to_excel("Weights Prediction.xlsx",header=True,index=True)
new_df.head()
```
